Uh-oh! It looks like we have some values in this column that are simply <code>-</code>, and Python can't convert these to a float type, which makes sense since they contain no numerical data. 

Let's take a look at every row with this value to confirm this is the case:

```python
ahl_2021_nhl_2122[ahl_2021_nhl_2122.ahl_ppg=="-"]
```

```python
# np.where syntax: np.where(condition==True, value_to_return_if_true, value_to_return_if_false)

ahl_2021_nhl_2122.ahl_ppg = np.where(ahl_2021_nhl_2122.ahl_ppg=="-", 0, ahl_2021_nhl_2122.ahl_ppg)
ahl_2021_nhl_2122.nhl_ppg = np.where(ahl_2021_nhl_2122.nhl_ppg=="-", 0, ahl_2021_nhl_2122.nhl_ppg)

ahl_2021_nhl_2122.ahl_gp = np.where(ahl_2021_nhl_2122.ahl_gp=="-", 0, ahl_2021_nhl_2122.ahl_gp)
ahl_2021_nhl_2122.nhl_gp = np.where(ahl_2021_nhl_2122.nhl_gp=="-", 0, ahl_2021_nhl_2122.nhl_gp)

ahl_2021_nhl_2122[ahl_2021_nhl_2122.ahl_ppg=="-"]
```

```python
ahl_2021_nhl_2122.ahl_ppg = ahl_2021_nhl_2122.ahl_ppg.astype(float) 

ahl_2021_nhl_2122.nhl_ppg = ahl_2021_nhl_2122.nhl_ppg.astype(float) 

ahl_2021_nhl_2122.ahl_gp = ahl_2021_nhl_2122.ahl_gp.astype(float) 

ahl_2021_nhl_2122.nhl_gp = ahl_2021_nhl_2122.nhl_gp.astype(float) 

ahl_2021_nhl_2122 = ahl_2021_nhl_2122.loc[(ahl_2021_nhl_2122.ahl_gp>=20) & (ahl_2021_nhl_2122.nhl_gp>=20)]

ahl_2021_nhl_2122
```

```python
# np.mean syntax: np.mean(array_of_values)

nhl_ppg_average = np.mean(ahl_2021_nhl_2122.nhl_ppg)

ahl_ppg_average = np.mean(ahl_2021_nhl_2122.ahl_ppg)

ahl_nhl_equivalency = nhl_ppg_average/ahl_ppg_average

ahl_nhl_equivalency
```

```python
# In line 1, we define our function as obtain_nhle_given_ahl_ppg and 1 argument to take in: ahl_ppg.
# In line 2, we calculate NHLe by multiplying AHL points per game by 82 and then multiplying that by the AHL to NHL equivalency we obtained previously.
# In line 3, we round this value to the 2nd decimal.
# In line 4, we print the NHLe.

def obtain_nhle_given_ahl_ppg(ahl_ppg):
    nhle = ahl_ppg * 82 * ahl_nhl_equivalency
    nhle = round(nhle, 2)
    print(nhle)
```
